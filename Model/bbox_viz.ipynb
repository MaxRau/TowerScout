```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
from glob import glob
```

```python
# read image
img = cv2.imread('data/nyc/1000002315.png')
```

```python
# read annotation file and loop over boxes

fn = glob('data/nyc_results/FP/1000002315*')


file1 = open(fn[0], 'r')
boxes = file1.readlines()
boxes

for b in boxes:
    [x,y,w,h] = [float(f) for f in b.split(" ")[1:5]]
    
    x1 = int(1280*(x-w/2))
    y1 = int(1280*(y-h/2))
    x2 = int(1280*(x+w/2))
    y2 = int(1280*(y+h/2))
    
    boxed = cv2.rectangle(img, (x1,y1), (x2,y2), (235, 52, 216), 2)

# x, y, w, h = 0.157421875, 0.771875, 0.04453125, 0.0546875
```

```python
plt.figure(figsize=(15,15))
plt.imshow(boxed)
plt.show()
```
